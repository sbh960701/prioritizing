```python
import pandas as pd
from sklearn.metrics import roc_curve
from sklearn.ensemble import IsolationForest
from sklearn import svm
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
#our system
df_our = pd.read_csv("tpr,fpr_our.csv", index_col = 0)

our_tpr = df_our["tpr"].values.tolist()
our_fpr = df_our["fpr"].values.tolist()
```

```python
#ds4n6
df_ds4n6 = pd.read_csv("tpr,fpr_ds4n6.csv", index_col = 0)

ds4n6_tpr = df_ds4n6["tpr"].values.tolist()
ds4n6_fpr = df_ds4n6["fpr"].values.tolist()
```

```python
#roc curve
plt.plot(our_fpr, our_tpr, '--', label='Our System(AUC=0.96)')
plt.plot(ds4n6_fpr, ds4n6_tpr, '--', label='DS4N6(AUC=0.79)')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.xlim(0, 1)

plt.legend()
plt.show()
```
